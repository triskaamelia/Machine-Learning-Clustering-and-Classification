```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
df = pd.read_csv('fix_air_bnb.csv')
df.head()
```

```python
df['kelayakan'] = np.random.randint(2, size=22467)
```

```python
df.kelayakan
```

```python
# split dataset menjadi 4 jenis dataset
# X_train : berisikan seluruh baris pada dataset yang bukan atribut kelas, digunakan sebagai data latih
# X_test : berisikan seluruh baris pada dataset yang bukan atribut kelas, digunakan sebagai data uji
# Y_train : berisikan seluruh baris pada dataset dengan atribut hanyalah atribut kelas, digunakan sebagai data latih kelas
# Y_test : berisikan seluruh baris pada dataset dengan atribut hanyalah atribut kelas, digunakan sebagai data uji kelas
X = df.iloc[:,[11,13]].values
Y = df.iloc[:,-1].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
# split lagi untuk membuat 2 jenis dataset, dimana dataset ini dapat digunakan untuk keperluan validasi
#X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2, random_state=1)
```

```python
# memasukan pengklasifikasi ke data latih
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,Y_train)
```

```python
# memprediksi nilai class milik semua data uji
Y_pred = classifier.predict(X_test)
```

```python
for i in Y_pred:
    print(i)
```

```python
# melihat keoptimalan classification dengan menggunakan confusion matrix
from sklearn.metrics import confusion_matrix, f1_score,precision_score,recall_score
import sklearn.metrics as skm
cm = confusion_matrix(Y_test, Y_pred)
print('confusion matrix')
print(cm)
print('')
conf = skm.multilabel_confusion_matrix(Y_test, Y_pred)
print('multilable confusion matrix')
print(conf)
print('')
print('--- Informasi Rumus Perhitungan ---')
print('accuracy = ((tp+tn)/(tp+tn+fp+fn))*100')
print('precision = (tp/(fp+tp))*100')
print('recall = (tp/(fn+tp))*100')
print('f1_score = 2* ((recall*presisi)/(recall+presisi))')
print('')
print( skm.classification_report(Y_test,Y_pred))
print('micro avg')
print('precision: ', precision_score(Y_test, Y_pred, average="micro"))
print('recall: ', recall_score(Y_test, Y_pred, average="micro"))
print('f1-score: ', f1_score(Y_test, Y_pred, average="micro"))
```

```python
# memvisualisasikan hasil data latih
from matplotlib.colors import ListedColormap
X_set, Y_set = X_train, Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.30, cmap = ListedColormap(('blue', 'orange','green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(Y_set)):
    plt.scatter(X_set[Y_set == j, 0], X_set[Y_set == j, 1],
                c = ListedColormap(('blue', 'orange','green'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('number_of_reviews')
plt.ylabel('reviews_per_month')
plt.legend()
plt.show()
```

```python
# memvisualisasikan hasil data uji
from matplotlib.colors import ListedColormap
X_set, Y_set = X_test, Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.30, cmap = ListedColormap(('blue', 'orange','green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(Y_set)):
    plt.scatter(X_set[Y_set == j, 0], X_set[Y_set == j, 1],
                c = ListedColormap(('blue', 'orange','green'))(i), label = j)
plt.title('Naive Bayes (Test set)')
plt.xlabel('number_of_reviews')
plt.ylabel('reviews_per_month')
plt.legend()
plt.show()
```
